```python
# importing numpy and pandas
# open up the dataset

import numpy as np
import pandas as pd

file = 'auto-new.csv'

cars = pd.read_csv('auto-new.csv')
```

```python
cars.head()
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
"drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
"num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
"peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
# inserting the headers to the data

cars.columns = headers
cars.head()
```

```python
# types of data

cars.dtypes
```

```python
# getting to know the size of the data

cars.shape
```

```python
# observing the data

cars.describe(include ='all')
```

```python
# dealing with the missing values

cars.replace('?', np.nan, inplace = True)
cars.head(7)

```

```python
#to find the number of missing values

missing_data = cars.isnull()
missing_data.head(10)
```

```python
#to count how many true and false in a column

for column in missing_data.columns.values.tolist():  
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
# get the mean of normalized-loss

mean_normloss = cars['normalized-losses'].astype('float').mean(axis=0)
print(mean_normloss)
```

```python
#replace current data with the mean value

cars['normalized-losses'].replace(np.nan, mean_normloss, inplace = True)
cars.head(5)

# as a result, row 0,1,4 are among the rows that have their values changed from nan to 122.
```

```python
# count the values in the column 

cars['make'].value_counts()
```

```python
# count the maximum value in the column

cars['price'].value_counts().idxmax()
```

```python
#replace current data with the value, formula: replace(current value, new value, inplace = True )

cars['num-of-doors'].replace(np.nan, 'four', inplace = True)
cars.head(7)

# row 27 for example, has its value of nan changed to four
```

```python
# dropping the rows of which price = true (missing value)

cars.dropna(subset = ['price'], axis=0, inplace = True)
```

```python
# showing the size of the table with price = True

cars.shape
```

```python
#the type result for 'price' came out as object instead of integer/float. To change it:

cars['price'] = cars['price'].astype('float')
```

```python
cars.dtypes
```

```python
# to create a new column based on the 'fuel-type' column (step 1)

dummyvar = pd.get_dummies(cars['fuel-type'])
dummyvar.head(7)
```

```python
#adding the above dataframe to the main data


cars = pd.concat([cars,dummyvar], axis=1)
cars.drop('fuel-type', axis = 1, inplace = True )
# we will see the diesel & fuel at the end of the table.
# fuel-type will be deleted.
```
